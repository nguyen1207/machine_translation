```python
from openai import OpenAI
import random
from random import shuffle

from tqdm import tqdm   

client = OpenAI(
    api_key = os.getenv('OPENAI_API_KEY')
)

def generate_example(prompt, examples, format_examples_func, temperature=0.7):
    messages = [
        {"role": "system", "content": "You are an expert in Vietnames lingustic having broad knowledge in Vietnamese vocabulary especially slang and teencode. You also have extensive knowledge in English vocabulary and grammar. Do your best to follow the instruction. Your response must be in JSON only."},
    ]
    
    k_shot_prompt = format_examples_func(prompt, examples)
    
    messages.append({
        "role": "user",
        "content": k_shot_prompt
    })
    

    response = client.chat.completions.create(
        model="gpt-4",
        messages=messages,
        temperature=temperature,
    )
    return response.choices[0].message.content.strip()
    
def format_example(prompt, examples):
    double_newline = "\n\n"
    newline = "\n"
    
    k_shot_prompt = \
f"""{prompt}

Here are {len(examples)} examples:

{double_newline.join(f"Example {i + 1}:{newline}Vietnamese: {example['vietnamese']}{newline}English: {example['english']}{newline}Correct alignments: {', '.join(example['alignments'])}{newline}Incorrect alignments: {', '.join(example['incorrect_alignments']) if 'incorrect_alignments' in example else ''}" for i, example in enumerate(examples))}

Please remember that your response must be in JSON only. Generate new records based on the given examples, do not include them in your response and try your best on using Vietnamese slang and teencode. I'm going to tip you $200 if you follow requirements and $200 more for each record that has Vietnamese slang and teencode.
"""
    
    return k_shot_prompt

prompt = \
"""You are generating data which will be used to train a LLM with the task machine translation from Vietnamese to English especially dealing with Vietnamese slang and teencode in informal tone. Here are the requirements:

1. You will be given a Vietnamese sentence and you need to translate it to English in informal tone. You can use slang and teencode in your translation. You will generate data samples, each with a Vietnamese sentence, its English translation and the words or short phrases alignments.

2. For each turn, make the example slightly more complex than the last, while ensuring diversity.

3. Make sure your samples are unique and diverse, yet high-quality and complex enough to train a well-performing machine translation model that can translate Vietnamese sentences to English in informal tone, especially dealing with Vietnamese slang and teencode. The model should be able to generate natural and fluent translations.

4. You must return a list of JSON object each entry must include 
- "vietnamse" which is the Vietnamese sentences (can be one or more)
- "english" which is the English translation of the previous Vietnames sentencse
- "alignments" contains a list of correct Vietnamese-English alignment word or short phrase pairs in format vi-en
- "incorrect_alignments" contains a list of incorrect alignments in format vi-en and must not be empty, the alignment must be completly wrong and only contain words or phrase in the sentence pair and must not be empty. Do not include punctuation.
- "fix" contains a list of fixes of incorrect alignments and must not be empty, the fixed alignment must only contain words or phrase in the sentence pair and must not be empty. Do not add random words to it even thought they may be correct. Do not include punctuation.

5. Do not generate too negative sentences or sentences that are too offensive or inappropriate. Keep the content positive and safe for all audiences.

6. Generate sentences pair that are in many different topics and contexts. It can be conversation, news, stories, etc. but make sure they are all in informal tone.
"""

# shuffle the data
storage = []
df_teencode = df_teencode.sample(frac=1).reset_index(drop=True)
df_nonteencode = df_nonteencode.sample(frac=1).reset_index(drop=True)

examples = []
generated = []

n = 30
start = 60

for t in tqdm(range(start, len(df_teencode), 2)):
    teencode = df_teencode.iloc[t:t+2].to_dict(orient='records')
    nonteencode = df_nonteencode.iloc[n:n+2].to_dict(orient='records')
    previous_generated_sample = []
    
    if len(generated) > 0:
        previous_generated_sample = random.sample(generated, 1)
        
    examples = nonteencode + previous_generated_sample + teencode
    shuffle(examples)
    
    while True:
        generated_examples = generate_example(prompt, examples, format_example)
        
        try:
            generated_examples_json = json.loads(generated_examples)
            
            generated.extend(generated_examples_json)
        except:
            continue
        
        break
    
    n += 1
    
```
